```
import pandas as pd
import numpy as np
data = pd.read_csv(r"./digit recognizor.csv")
X = data.iloc[:,1:]
y = data.iloc[:,0]

print(X.shape)
```

确定嵌入法选择特征的个数

```
from sklearn.feature_selection import SelectFromModel
from sklearn.ensemble import RandomForestClassifier as RFC
RFC_ = RFC(n_estimators =10,random_state=0)
X_embedded = SelectFromModel(RFC_,threshold=0.0045).fit_transform(X,y)
```

```
X_embedded.shape
```

过滤法选择特征

```
xx = X.var().values
```

```
xx
```

```
index = xx.argsort()
```

```
index
```

```
th = xx[index[-62]]
th
```

```
from sklearn.feature_selection import VarianceThreshold
X_fsvar = VarianceThreshold(th).fit_transform(X)
X_fsvar.shape
```

包装法

```
from sklearn.feature_selection import RFE
selector = RFE(RFC_, n_features_to_select=61, step=50).fit(X, y)
X_wrapper = selector.transform(X)
X_wrapper.shape
```

## 结果对比

```
from sklearn.model_selection import cross_val_score
cross_val_score(RFC_,X_fsvar,y,cv=5).mean()
```

```
cross_val_score(RFC_,X_embedded,y,cv=5).mean()
```

```
cross_val_score(RFC_,X_wrapper,y,cv=5).mean()
```

```
cross_val_score(RFC(n_estimators=100,random_state=0),X_embedded,y,cv=5).mean()
```

```
cross_val_score(RFC(n_estimators=100,random_state=0),X_wrapper,y,cv=5).mean()
```

```
cross_val_score(RFC(n_estimators=100,random_state=0),X_fsvar,y,cv=5).mean()
```

## 特征更少时的对比

```
from sklearn.feature_selection import SelectFromModel
from sklearn.ensemble import RandomForestClassifier as RFC
RFC_ = RFC(n_estimators =10,random_state=0)
X_embedded2 = SelectFromModel(RFC_,threshold=0.005).fit_transform(X,y)
X_embedded2.shape
```

```
from sklearn.feature_selection import RFE
selector2 = RFE(RFC_, n_features_to_select=47, step=50).fit(X, y)
X_wrapper2 = selector2.transform(X)
X_wrapper2.shape
```

```
cross_val_score(RFC_,X_embedded2,y,cv=5).mean()
```
